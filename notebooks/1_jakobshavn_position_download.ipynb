## Download MEaSUREs-0781 for Jakobshavn Isbrae ##

This notebook uses already downloaded data from the [MEaSUREs Weekly to Monthly Greenland Outlet Glacier Terminus Positions from Sentinel-1 Mosaics](https://nsidc.org/data/nsidc-0781/versions/1)

```python
#import libraries and packages
import earthaccess
import geopandas as gpd
from pathlib import Path 
import os
```

```python
data_path = Path("/Users/sofiasuhinin/Desktop/GrIS_Calving/MEaSUREs_0781")
termini = gpd.read_file(data_path / "glacier_termini_v01.0.shp")
points = gpd.read_file(data_path / "glacier_points_v01.0.shp")
boxes = gpd.read_file(data_path / "glacier_boxes_v01.0.shp")
```

```python
print(points.loc[points["Glacier_ID"]==3, ["Official_N","Alternativ","Glacier_ID"]])

# subset the termini data to grab specifically Jakobshavn Isbrae (Glacier ID 3)
jakobshavn = termini[termini["Glacier_ID"] == 3]
```

```python
outdir = Path("/Users/sofiasuhinin/Desktop/GrIS_Calving/MeASUREs_0781/Jakobshavn_Isbrae")
outdir.mkdir(exist_ok=True, parents=True) #create directory if it doesn't exist

output_file = outdir / "JakbobshavnIsbraeTerminus.shp"
jakobshavn.to_file(output_file)
```
